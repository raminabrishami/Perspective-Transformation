```python
from skimage import io
import matplotlib.pyplot as plt
import cv2
import numpy as np
import ast
```

```python
image=input('lotfan address file ra vared konid : ')
location=ast.literal_eval(input('lotfan location aks ra vared konid ba format mesal [10,56],[26,93],[67,258],[234,300] : '))
size_image=ast.literal_eval(input('lotfan size aks nahayi ra vared konid ba format mesal [0,0],[300,0],[0,300],[300,300] : '))
s=size_image[3][1]

def Transformation(img,location,size_image):
    img = io.imread(img)
    rows,cols,ch = img.shape
    pts1 = np.float32([location])
    pts2 = np.float32([size_image])
    M = cv2.getPerspectiveTransform(pts1,pts2)
    dst = cv2.warpPerspective(img,M,(s,s))
    plt.subplot(121),plt.imshow(img),plt.title('Input')
    plt.subplot(122),plt.imshow(dst),plt.title('Output')
    plt.show()
```

```python
Transformation(image,location,size_image)
```
